```python
import numpy as np
```

```python
K_sim_train = np.load("kernel_matrix_sim.npy")
K_quantum_train = np.load("kernel_matrix_quantum.npy")
```

```python
fro_diff = np.linalg.norm(K_sim_train - K_quantum_train, 'fro')
fro_sim = np.linalg.norm(K_sim_train, 'fro')

relative_error = fro_diff / fro_sim
print("Relative Frobenius error:", relative_error)
```
