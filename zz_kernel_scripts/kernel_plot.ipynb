```python
import matplotlib.pyplot as plt
import numpy as np

#200 DATA POINT

#Precomputed kernel classification test score: 0.5087719298245614
#Linear SVM accuracy: 0.43859649122807015
#
#Precomputed kernel classification test score: 0.47368421052631576
#Linear SVM accuracy: 0.45614035087719296
#
#Precomputed kernel classification test score: 0.5263157894736842
#Linear SVM accuracy: 0.43859649122807015
#RBF SVM accuracy: 0.5087719298245614
#
#Precomputed kernel classification test score: 0.5964912280701754
#Linear SVM accuracy: 0.38596491228070173
#RBF SVM accuracy: 0.47368421052631576
#Precomputed kernel classification test score: 0.45614035087719296
#Linear SVM accuracy: 0.3508771929824561
#RBF SVM accuracy: 0.5614035087719298


f1_scores_linear = [0.43859649122807015, 0.45614035087719296, 0.43859649122807015 , 0.38596491228070173, 0.3508771929824561]
f1_scores_rbf = [0.5087719298245614, 0.47368421052631576, 0.5614035087719298]
f1_scores_quantum = [0.5087719298245614, 0.47368421052631576, 0.5263157894736842, 0.5964912280701754, 0.45614035087719296]
import matplotlib.pyplot as plt
import numpy as np

f1_scores_linear = [0.4386, 0.4561, 0.4386, 0.3860]
f1_scores_rbf = [0.5088, 0.4737]
f1_scores_quantum = [0.5088, 0.4737, 0.5263, 0.5965]

data = [f1_scores_linear, f1_scores_rbf, f1_scores_quantum]
x_pos = [0, 1, 2]

plt.figure(figsize=(8,6), facecolor='none')

box = plt.boxplot(
    data,
    positions=x_pos,
    patch_artist=True,
    widths=0.25,
    showfliers=False,
    showmeans=True,
)

# Transparent box interiors with thick edges
colors = ['blue', 'red', 'darkorange']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor('none')

# Customize whiskers, caps, and median
for whisker in box['whiskers']:
    whisker.set_color('black')
for cap in box['caps']:
    cap.set_color('black')
for median in box['medians']:
    median.set_color('black')



# Overlay points
jitter = 0.06
plt.scatter(np.random.normal(0, jitter, len(f1_scores_linear)), f1_scores_linear, color='blue', alpha=0.6)
plt.scatter(np.random.normal(1, jitter, len(f1_scores_rbf)), f1_scores_rbf, color='red', alpha=0.6)
plt.scatter(np.random.normal(2, jitter, len(f1_scores_quantum)), f1_scores_quantum, color='darkorange', alpha=0.6)

plt.xticks(x_pos, ['Linear SVM', 'RBF SVM', 'Quantum kernel'])
plt.ylabel('F1 score')
plt.title('F1-score distributions (random train/test splits)')
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.legend(framealpha=0.0)

plt.gca().patch.set_alpha(0.0)
plt.savefig('f1_boxplot_kernel.png', transparent=True)
plt.show()
```
